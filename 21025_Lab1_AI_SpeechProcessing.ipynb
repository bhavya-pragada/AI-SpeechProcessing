```python
import librosa
import librosa.display
import IPython.display as ipd
from IPython.display import Audio
```

```python
ipd.Audio('bhavya_voice1.wav')
```

```python
y, sr = librosa.load('bhavya_voice1.wav')
```

```python
librosa.display.waveshow(y)
```

```python
import numpy as np
```

### A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.

```python
voice_len = len(y)
print("Length of the signal: ",voice_len)

voice_magnitude = np.max(y), np.min(y)
print("Magnitude range: ",voice_magnitude)
```

### A3. Take a small segment of the signal and play it.

```python
start_time = 0.5
end_time = 2.0
segment1 = y[int(start_time * sr):int(end_time * sr)]
Audio(segment1, rate=sr)
```

```python
librosa.display.waveshow(segment1)
```
