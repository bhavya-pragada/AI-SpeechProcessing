### A1.Findthe first derivative of your speech signal with finite differencemethod. Listen to the first derivative signal and the original speech signal.

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from IPython.display import Audio, display
```

```python
voice1,fs=librosa.load('Bhavya_voice1.wav')
```

```python
print("Original Speech Signal")
```

```python
ipd.Audio('Bhavya_voice1.wav')
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(voice1)
plt.title("Original Speech Signal")
plt.show()
```

```python
first_deriv= np.concatenate(([0], np.diff(voice1, n=1)))
```

```python
Audio(data=first_deriv,rate=fs)
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(first_deriv, sr=fs)
plt.title("First Derivative Signal")
plt.show()
```

### A2. Detect the points of zero crossing in the first derivative signal. Compare the average length between two consecutive zero crossings for speech and silence regions. Observe the pattern.

```python
zero_crossings = np.where(np.diff(np.sign(first_deriv)))[0]
crossing_lengths = np.diff(zero_crossings)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(first_deriv)
plt.scatter(zero_crossings, first_deriv[zero_crossings], c='r', marker='x', label='Zero Crossing')
plt.title("First Derivative Signal with Zero Crossings")
plt.legend()
plt.show()
```

```python
# average length between two consecutive zero crossings
```

```python
avg_len = np.diff(zero_crossings)
speech_length = avg_len[:len(avg_len)//2]
silence_length = avg_len[len(avg_len)//2:]
```

```python
avg_speech = np.mean(speech_length)
avg_silence = np.mean(silence_length)
print("Average length b/w two consecutive zero crossings for speech: ",avg_speech)
print("Average length b/w two consecutive zero crossings for silence: ",avg_silence)
```
